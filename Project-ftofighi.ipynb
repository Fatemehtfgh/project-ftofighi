## Part 1: Initial Setup
Set up the foundational structure for your weekly task and habit tracker.
### Instructions:
1. Create a list containing all the days of the week **('Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday')**.
2. Initialize an empty dictionary that will be used to track habits for the week.
3. Create a second dictionary that uses the days of the week as keys, with each key initially linked to an empty dictionary for storing tasks.

```python
# list of the days of the week
days_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Dictionary to track habits
habit_tracker = {}

# Dictionary for storing tasks
task_tracker = {
    "Monday": {},
    "Tuesday": {},
    "Wednesday": {},
    "Thursday": {},
    "Friday": {},
    "Saturday": {},
    "Sunday": {}
}
```

## Part 2: Add Tasks and Habits
Build functions that let users add tasks to specific days and habits for the entire week.
### Instructions
1. **Create a function to add tasks:**
- Ask the user which day they want to add the task to.
- Make sure to validate the day. Display a message if the day was invalid.
- Ask for the task name.
- Store it in the dictionary you created to store tasks under the correct day.
- Set the task’s value to *False* to show it hasn’t been completed yet.


2. **Create a function to add habits:**
- Ask the user for the name of the habit (e.g., “Drink Water”).
- Make sure to check if the habit already exists. Display a message if it does.
- If it’s new, add it to the dictionary you created to store habits.
- For that habit, create a nested dictionary with all 7 days set to *False*.


**Tip:** Use ***.strip()*** and ***.title()*** to clean up user input.
- **.strip()** = Removes any extra spaces at the beginning or end of the text. People often hit the spacebar by accident and add leading or trailing unwanted spaces. Example: "  Monday  " can be fixed by using .strip(): "Monday""
- **.title()** = Capitalizes the first letter of each word and makes the rest lowercase. It helps match formatting exactly (especially for days of the week). Example: "monday" or "MONDAY"" will be changed to "Monday".
- **If the day does not exist, show a helpful error message.**

```python
# function to add tasks
def add_task(task_tracker):
    day = input("Enter the day you want to add the task to: ").strip().title()
    if day not in days_week:
        print("That's not a valid day. Try again.")
        return
    task = input("Enter the task name: ").strip()
    if task in task_tracker[day]:
        print("Task already exists for this day.")
    else:
        task_tracker[day][task] = False # Not completed yet
        print(f"Task '{task}' added to {day}.")
        
#add_task(task_tracker)     
```

```python
# function to add habit
def add_habit(habit_tracker):
    habit = input("Enter the name of the habit: ").strip().title()
    if habit in habit_tracker:
        print("Habit already exists.")
    else:
        # make habit trackable
        habit_tracker[habit] = {
            'Monday': False,
            'Tuesday': False,
            'Wednesday': False,
            'Thursday': False,
            'Friday': False,
            'Saturday': False,
            'Sunday': False
        }
        print(f"Habit '{habit}' added for all days.")
        
#add_habit(habit_tracker)
```

## Part 3: Marking Tasks and Habits as Complete
In this part, you'll implement two functions to allow users to mark their progress: one for tasks and one for habits.
### Instructions:
1. Create a function to mark tasks as complete:
- Display all uncompleted tasks along with the day they belong to.
- Ask the user to enter the name of the task they completed.
- Find and update the task in your data structure by setting its value to *True*.
- If all tasks are completed, display a message saying so.
- Make sure to validate the day and check if the task exists. Display a message if the task does not exist or if the day was invalid.
- **BONUS:** If the task is from a day before today, print a encouraging or funny message like: "Task marked as complete. Better late than never!" **Tip:** You can use the *datetime* library.

2. Create a function to mark habits as complete:
- Ask the user to enter the day and the name of the habit.
- Make sure to validate the day and check if the habit exists. Display a message if the habit does not exist or if the day was invalid.
- Update the habit’s status for that day by setting its value to *True*.

#### Tips:
- Use **.strip()** and **.title()** to clean up user input.
- Always check if the dictionary contains data before trying to access something.
- **If the day, habit or task doesn't exist, show a helpful error message.**
  
**Helpful note:** You can access a dictionary’s information using loops like this:

```python
my_dict = {
    "a": 1,
    "b": 2,
    "c": 3
}

# Loop through keys and values
for key, value in my_dict.items():
    print(f"{key} = {value}")
```

This will print each key and its value.

You can also loop just through keys with ***for key in my_dict:*** or just values with for value in ***my_dict.values():***

```python
# Create a function to mark tasks as complete
import datetime
#Display all uncompleted tasks along with the day
def completed_task(task_tracker):
    found = False
    for day in task_tracker:
        for task, done in task_tracker[day].items():
            if not done:
                print(f"- {task} ({day})")
                found = True
    if not found:
        print("All tasks are completed!")
        return

    # Ask the user for the day
    day = input("Which day is the task for? ").strip().title()
    if day not in task_tracker:
        print("That’s not a valid day. Please try again.")
        return

    # Ask for the task name
    task_name = input("What task did you complete? ").strip()

    # Check if the task exists on that day
    if task_name not in task_tracker[day]:
        print(f"Task '{task_name}' was not found on {day}.")
        return

    # Mark the task as complete
    task_tracker[day][task_name] = True
    print(f" Task '{task_name}' marked as complete for {day}.")

    # BONUS: encouraging message if it's late
    today = datetime.datetime.today().strftime('%A')
    if days_week.index(day) < days_week.index(today):
        print("Task marked as complete. Better late than never!")
    else:
        print("Task marked as complete. Good job!")
        
#completed_task(task_tracker)
```

```python
#Create a function to mark habits as complete

def completed_habit(habit_tracker):
    # Ask the user for the day
    day = input("Which day did you complete the habit? ").strip().title()

    if day not in days_week:
        print("That’s not a valid day. Please try again.")
        return

    # Ask for the habit name
    habit = input("What habit did you complete? ").strip().title()

    if habit not in habit_tracker:
        print("That habit is not being tracked.")
        return

    # Mark the habit as completed for that day
    habit_tracker[habit][day] = True
    print(f" Habit '{habit}' marked as complete for {day}.")

#completed_habit(habit_tracker)
```

```python
#Create a function to remove a task
def remove_task(task_tracker):
    day = input("Which day is the task on? ").strip().title()

    if day not in task_tracker:
        print("That’s not a valid day.")
        return

    if not task_tracker[day]:
        print(f"There are no tasks for {day}.")
        return

    print(f"Tasks on {day}:")
    for task in task_tracker[day]:
        print(f"- {task}")

    # Ask the user which task to remove
    task_name = input("Which task do you want to remove? ").strip()

    if task_name in task_tracker[day]:
        del task_tracker[day][task_name]
        print(f"Task '{task_name}' removed from {day}.")
    else:
        print("Task not found.")

#remove_task(task_tracker)
```

```python
#Create a function to remove a habit
def remove_habit(habit_tracker):
    if not habit_tracker:
        print("You don’t have any habits to remove.")
        return

    print("Your habits:")
    for habit in habit_tracker:
        print(f"- {habit}")

    #Ask the user which habit to remove
    habit_name = input("Which habit do you want to remove? ").strip().title()

    if habit_name in habit_tracker:
        del habit_tracker[habit_name]
        print(f"Habit '{habit_name}' has been removed.")
    else:
        print("Habit not found.")

#remove_habit(habit_tracker)
```

## Part 5: Generating Weekly and Daily Reports

### Instructions:
1. Create a Weekly Report Function

This function should summarize the user's progress over the entire week.
- For Habits:
    - Loop through each habit in your habit tracker.
    - Count how many days the habit was marked as complete using a loop or a list comprehension.
    - Display the habit name and the number of days it was completed out of 7.
    - Display a message if there are no habits found.
- For Tasks:
    - Loop through all days and collect tasks into two separate lists:
        - One for completed tasks
        - One for not completed tasks
    - Include the task name and the day in parentheses.
    - Display both lists clearly.
    - Display a message if there are no tasks found.

2. Create a Daily Report Function

This function allows the user to view all their activity for a specific day.

- Ask the user to enter a day (e.g., "Monday").
- If the day is valid:
    - Display all tasks for that day, showing whether each one is complete or not.
    - Display all habits, showing whether they were completed on that day.
- Use ✅ for completed and ❌ for not completed.

#### Tips:
- Use **.title()** to make sure the input day matches the format in your list.
- Use clear formatting in your print statements to separate sections.
- Always check if the dictionary contains data before trying to access something.
- **If the day doesn't exist, show a helpful error message.**

```python
#summarize the user's progress over the entire week
#For Habits
def habit_progress(habit_tracker):
    print("\nHabit progress:")

    if not habit_tracker:
        print("There are no habits found")
        return

    for habit in habit_tracker:
        total_days = len(habit_tracker[habit])
        completed_days = 0

        for day in habit_tracker[habit]:
            if habit_tracker[habit][day] == True:
                completed_days += 1

        print(f"{habit}: {completed_days}/{total_days} days completed")

#For Tasks
def task_progress(task_tracker):
    print("\nTask progress:")

    completed_tasks = []
    uncompleted_tasks = []

    # Loops through each day and each task within that day
    for day in task_tracker:
        for task in task_tracker[day]:
            if task_tracker[day][task] == True:
                completed_tasks.append(task + " (" + day + ")")
            else:
                uncompleted_tasks.append(task + " (" + day + ")")

    # If both lists are empty, no tasks have been added yet
    if len(completed_tasks) == 0 and len(uncompleted_tasks) == 0:
        print("There are no tasks found.")
        return

    # Show uncompleted tasks
    print("\nUncompleted Tasks:")
    if len(uncompleted_tasks) == 0:
        print("All tasks are completed.")
    else:
        for task in uncompleted_tasks:
            print(task)

    # Show completed tasks
    print("\nCompleted Tasks:")
    if len(completed_tasks) == 0:
        print("No tasks completed yet.")
    else:
        for task in completed_tasks:
            print(task)

#habit_progress(habit_tracker)
#task_progress(task_tracker)
```

```python
#Create a Daily Report Function
def daily_report(task_tracker, habit_tracker):
    # Ask the user to enter a day
    day = input("Enter the day you want to view all your activity: ").strip().title()

    if day not in days_week:
        print("That’s not a valid day. Please try again.")
        return

    print(f"\n Daily Report for {day}")

    print("\n Tasks:")

    if day not in task_tracker or not task_tracker[day]:
        print("No tasks found for this day.")
    else:
        for task in task_tracker[day]:
            status = "✅" if task_tracker[day][task] else "❌"
            print(f"{status} {task}")

    print("\n Habits:")

    if not habit_tracker:
        print("No habits are being tracked.")
    else:
        habits_found = False
        for habit in habit_tracker:
            if day in habit_tracker[habit]:
                status = "✅" if habit_tracker[habit][day] else "❌"
                print(f"{status} {habit}")
                habits_found = True

        if not habits_found:
            print("No habit found for this day.")

#daily_report(task_tracker, habit_tracker)
```
